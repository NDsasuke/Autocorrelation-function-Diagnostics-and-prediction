1. Import the necessary libraries:
   - `numpy` for numerical operations.
   - `pandas` for handling time series data.
   - `matplotlib.pyplot` for data visualization.
   - `statsmodels.graphics.tsaplots.plot_acf` and `statsmodels.graphics.tsaplots.plot_pacf` for computing and plotting the ACF and PACF respectively.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

2. Load the time series data:
   - Replace `'your_dataset.csv'` with the path or filename of your own dataset.
   - The data should be in a CSV format with a column representing the dates and another column representing the values of the time series.
   - The `parse_dates` argument is used to parse the dates column as datetime objects.
   - The `index_col` argument is used to set the dates column as the index of the DataFrame.



```python
# Load the Air Passengers dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
```

3. Plot the original time series:
   - Creates a new figure with a size of 10x5 using `plt.figure(figsize=(10, 5))`.
   - Plots the time series data using `plt.plot()`.
   - Adds x-axis label, y-axis label, and title using `plt.xlabel()`, `plt.ylabel()`, and `plt.title()` respectively.
   - Gridlines are added to the plot using `plt.grid(True)`.
   - Displays the plot using `plt.show()`.



```python
# Plot the original time series
plt.figure(figsize=(10, 5))
plt.plot(data)
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Original Time Series')
plt.grid(True)
plt.show()
```

4. Compute and plot the ACF:
   - Creates a new figure with a size of 10x5 using `plt.figure(figsize=(10, 5))`.
   - Uses `plot_acf()` from `statsmodels.graphics.tsaplots` to compute and plot the ACF.
   - The `lags` parameter specifies the number of lags to be included in the ACF plot.
   - The `alpha` parameter controls the confidence interval for the plot.
   - Adds x-axis label, y-axis label, and title using `plt.xlabel()`, `plt.ylabel()`, and `plt.title()` respectively.
   - Gridlines are added to the plot using `plt.grid(True)`.
   - Displays the plot using `plt.show()`.



```python
# Compute and plot the ACF
plt.figure(figsize=(10, 5))
plot_acf(data, lags=30, alpha=0.05)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function (ACF)')
plt.grid(True)
plt.show()
```

5. Compute and plot the PACF:
   - Creates a new figure with a size of 10x5 using `plt.figure(figsize=(10, 5))`.
   - Uses `plot_pacf()` from `statsmodels.graphics.tsaplots` to compute and plot the PACF.
   - The `lags` parameter specifies the number of lags to be included in the PACF plot.
   - The `alpha` parameter controls the confidence interval for the plot.
   - Adds x-axis label, y-axis label, and title using `plt.xlabel()`, `plt.ylabel()`, and `plt.title()` respectively.
   - Gridlines are added to the plot using `plt.grid(True)`.
   - Displays the plot using `plt.show()`.


```python
# Compute and plot the PACF
plt.figure(figsize=(10, 5))
plot_pacf(data, lags=30, alpha=0.05)
plt.xlabel('Lag')
plt.ylabel('Partial Autocorrelation')
plt.title('Partial Autocorrelation Function (PACF)')
plt.grid(True)
plt.show()
```
